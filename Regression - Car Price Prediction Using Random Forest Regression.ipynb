```python
df=df.drop(['car_ID'],axis=1)
df.head()
```

## Splitting data into Independent and Dependent variable

```python
# In this model we are going to use only numerical values to predict the price of a car
X=df.iloc[:,:-1]
numericalCols=X.select_dtypes(exclude=['object']).columns
X=X[numericalCols]
X
```

```python
Y=df.iloc[:,-1].values
Y
```

## Scaling the Features

```python
from sklearn.preprocessing import scale
```

```python
cols=X.columns
X=pd.DataFrame(scale(X))
X.columns=cols
X
```

## Splitting of a dataset into Training and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=0)
X_train,X_test,Y_train,Y_test
```

## Train the model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
model.fit(X_train,Y_train)
```

## Prediction

```python
Y_pred=model.predict(X_test)
Y_pred
```

## Accuracy

```python
from sklearn.metrics import r2_score
```

```python
r2score=r2_score(Y_test,Y_pred)
r2score*100
```
