# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
df=datasets.load_iris()
```

```python
df
```

## Data Segregation

```python
X=df.data
```

```python
X
```

```python
Y=df.target
```

```python
Y
```

```python
names=df.target_names
names
```

## Fitting the PCA clustering to the dataset with n=2

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)    #Number of components to keep
Y_means=model.fit(X).transform(X)
```

```python
Y_means
```

## Variance Percentage

```python
plt.figure()
colors=['red','green','orange']
for color,i,target_name in zip(colors,[0,1,2],names):
    plt.scatter(Y_means[Y==i,0],Y_means[Y==i,1],color=color,lw=2,label=names)
plt.title('Iris Clustering')
plt.legend()
```
